```python
import numpy as np
import matplotlib as plt
import torch
import torchvision
import torch.nn as nn
import torchvision.transforms as transforms

```

```python
input_size = 784
num_classes = 10
num_epochs = 5
alpha = 0.002
batch_size = 100
```

```python
train_dataset = torchvision.datasets.MNIST(root='./data',train=True,transform=transforms.ToTensor(),download=True) 
                                           
                                         
                                          

test_dataset = torchvision.datasets.MNIST(root='./data',train=False,transform=transforms.ToTensor())
train_loader = torch.utils.data.DataLoader(dataset=train_dataset, 
                                           batch_size=batch_size, 
shuffle=True)
test_loader = torch.utils.data.DataLoader(dataset=test_dataset, 
                                          batch_size=batch_size, 
shuffle=False)
                                           

```

```python
model=nn.Linear(input_size,num_classes)
criterion=nn.CrossEntropyLoss()#logistic regression loss function
optimizer=torch.optim.SGD(model.parameters(),lr=alpha)
#print(train_loader)
count=len(train_loader)
for epoch in range(num_epochs):
    for i,(images,labels) in enumerate(train_loader):
        images=images.reshape(-1,28*28)#reshape into batch_size by 784
        outputs=model(images)
        loss=criterion(outputs,labels)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        if (i+1) % 100 == 0:
            print ('Epoch [{}/{}], Step [{}/{}], Loss: {:.4f}' 
                   .format(epoch+1, num_epochs, i+1, count
                           , loss.item()))

```

```python
with torch.no_grad():#sets all requires_grad to false temporarily as we don't need to compute gradients for test data
    total=0
    correct=0
    for images,labels in test_loader:
        images=images.reshape(-1,784)
        outputs=model(images)
        _,predicted = torch.max(outputs,1)#finds max of outputs along each row and returns both maximum value and position
        total+=labels.size(0)
        correct+=(labels==predicted).sum()
        print('Accuracy of the model on the 10000 test images: {} %'.format(100 * correct / total))
        
```
